```python
import annotations
full_annotator = annotations.FullAnnotator().add_all()
print("Total number of annotators:", len(full_annotator.annotators))
full_annotator.annotate_docbin("./data/conll2003bio.docbin")
```

```python
import labelling3

sources_to_use = [l for l in labelling3.SOURCE_NAMES if "conll2003" not in l]
unified_model = labelling3.HMMAnnotator(sources_to_use)

# We then run Baum-Welch on the model (can take some time)
unified_model.train("./data/conll2003bio.docbin")

# Saving the model to a file
unified_model.save("./data/hmm_conll2003bio.pkl")

unified_model.annotate_docbin("./data/conll2003bio.docbin")
```

```python
import annotations, analysis
docs = list(annotations.docbin_reader("./data/conll2003bio.docbin"))
print(len(docs))
df = analysis.evaluate(docs, ["HMM"], analysis.CONLL_MAPPINGS, analysis.CONLL_TO_RETAIN)
df.loc["micro"]
```
